```python
import numpy as np
import matplotlib.pyplot as plt
```

## Funcion de activacion: Sigmoid

_Formula:_ $$\sigma(x) = \frac{1}{1 + e^{-x}}$$
[Wolfram Alpha](https://www.wolframalpha.com/input?i=sigmoid%28x%29&lang=es)

La función sigmoid tiene una forma de "S" y se acerca asintóticamente a 1 en el lado positivo del eje y y asintóticamente a 0 en el lado negativo del eje y. La función tiene un valor de salida en el rango de 0 a 1, lo que la hace útil para problemas de clasificación binaria.

En una red neuronal, la función sigmoid se utiliza comúnmente en la capa de salida para producir la probabilidad de que la entrada pertenezca a la clase positiva (por ejemplo, la probabilidad de que una imagen sea un perro en lugar de un gato).

Sin embargo, la función sigmoid tiene una desventaja conocida como el problema de la desaparición del gradiente, que ocurre cuando los valores de entrada son muy grandes o muy pequeños. En estos casos, los gradientes de la función sigmoid son muy pequeños, lo que puede hacer que el proceso de aprendizaje sea lento o incluso imposible en algunas situaciones. Debido a esta desventaja, la función sigmoid ha sido en gran parte reemplazada por otras funciones de activación como la ReLU (Rectified Linear Unit).

```python
sigmoid = lambda x: 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(10, -10, 100)
x
```

```python
plt.plot(x, sigmoid(x))
plt.title('Función de activación sigmoid')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.show()
```

## Funcion de activacion: escalonada/step/threshold

_Formula:_ $$\sigma(x) = \begin{cases} 0 & \text{si } x < 0 \\ 1 & \text{si } x \geq 0 \end{cases}$$
[Wolfram Alpha](https://www.wolframalpha.com/input?i=step%28x%29&lang=es)

la función step devuelve 0 si la entrada es negativa y 1 si la entrada es no negativa.

La función step es una función de umbral, lo que significa que la salida cambia de un valor a otro en un punto de umbral. En una red neuronal, la función step se utiliza comúnmente en la capa de salida para producir una salida binaria, como en el caso de un clasificador de dos clases.

Aunque la función step es útil en algunos casos, también tiene una desventaja importante. En particular, la función step no es diferenciable en el punto de umbral. Esto significa que el algoritmo de optimización utilizado para entrenar la red neuronal puede tener dificultades para ajustar los pesos de las conexiones entre neuronas, lo que puede hacer que el proceso de aprendizaje sea lento o incluso imposible en algunas situaciones. Debido a esta desventaja, la función step no se utiliza comúnmente en la mayoría de las aplicaciones de redes neuronales modernas.

```python
step = lambda x: np.where(x >= 0, 1, 0)
```

```python
plt.plot(x, step(x))
plt.title('Función de activación step')
plt.xlabel('x')
plt.ylabel('step(x)')
plt.show()
```

## Funcion de activacion: ReLU

_Formula:_ $$\sigma(x) = \begin{cases} 0 & \text{si } x < 0 \\ x & \text{si } x \geq 0 \end{cases}$$
[Wolfram Alpha](https://www.wolframalpha.com/input?i=relu%28x%29&lang=es)

la función ReLU devuelve 0 para cualquier valor de x que sea menor que 0, y devuelve x para cualquier valor de x que sea mayor o igual a 0. Esto significa que la función es lineal para valores positivos de x y cero para valores negativos de x.

La función ReLU se utiliza comúnmente en redes neuronales para mejorar el rendimiento del modelo en problemas de clasificación y regresión. En particular, la función ReLU se ha demostrado que reduce el problema de la desaparición del gradiente en comparación con funciones de activación como la sigmoid o la tangente hiperbólica. La función ReLU también es computacionalmente eficiente y fácil de calcular, lo que la hace ideal para su uso en redes neuronales profundas.

Sin embargo, la función ReLU también tiene una desventaja conocida como "dying ReLU problem", que ocurre cuando algunos de los valores de x son muy pequeños y, por lo tanto, la función ReLU se satura en cero. Cuando esto ocurre, las neuronas afectadas por este problema ya no pueden actualizar sus pesos durante el entrenamiento, lo que puede hacer que la red neuronal sea menos efectiva en la tarea que se le ha asignado. Para solucionar este problema, se han propuesto variantes de la función ReLU, como la Leaky ReLU y la ELU.


```python
ReLU = lambda x: np.where(x >= 0, x, 0)
```

```python
plt.plot(x, ReLU(x))
plt.title('Función de activación ReLU')
plt.xlabel('x')
plt.ylabel('ReLU(x)')
plt.show()
```

## Funcion de activacion: tanh

_Formula:_ $$\sigma(x) = \frac{senh(x)}{cosh(x)} = \frac{e^x - e^{-x}}{e^x + e^{-x}}$$
[Wolfram Alpha](https://www.wolframalpha.com/input?i=tanh%28x%29&lang=es)

la función tanh es una versión escalada y desplazada de la función sigmoid, que devuelve valores en el rango de -1 a 1. La función tanh es simétrica en torno al origen y tiene una forma de "S" similar a la función sigmoid.

La función tanh se utiliza comúnmente en redes neuronales para introducir no linealidad en el modelo. La función tanh es útil para problemas de clasificación binaria, en los que la salida de la red neuronal debe ser una probabilidad en el rango de 0 a 1. La función tanh también se utiliza a menudo en capas ocultas de redes neuronales para suavizar la salida de la capa anterior y permitir una mejor propagación de la señal a través de la red neuronal.

Sin embargo, la función tanh también tiene algunas desventajas, como el problema de la desaparición del gradiente y la no linealidad que puede ser excesiva para algunos problemas de modelado. Por lo tanto, en algunos casos, se prefieren otras funciones de activación, como la ReLU o la Leaky ReLU.

```python
tanh = lambda x: (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
```

```python
plt.plot(x, tanh(x))
plt.title('Función de activación tanh')
plt.xlabel('x')
plt.ylabel('tanh(x)')
plt.show()
```
